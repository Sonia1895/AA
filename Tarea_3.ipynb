#Comprobación de datos paramétricos o no paramétrico

```python
from scipy.stats import normaltest


# Filtrar variables que tienen suficientes valores únicos como para asumir que son continuas
min_unique = 5

continuous_vars = [col for col in df_datos_relacion.columns if df_datos_relacion[col].nunique() >= min_unique]

print("Variables consideradas continuas para test de normalidad:")
print(continuous_vars)
print('-------------------'*5)
# Diccionario para guardar resultados
normality_results = {}

for col in continuous_vars:
    # Eliminar NaNs
    data = df_datos_relacion[col].dropna()

    # Evitar fallos en columnas constantes o muy pequeñas
    if len(data) >= 8 and data.nunique() > 1:
        # D’Agostino y Pearson test
        stat, p = normaltest(data)
        normality_results[col] = p
    else:
        normality_results[col] = None

# Mostrar resultados
for var, p_value in normality_results.items():
    if p_value is None:
        print(f"{var}: No aplicable")
    elif p_value > 0.05:
        print(f"{var}: ✅ Paramétrico (p = {p_value:.4f})")
    else:
        print(f"{var}: ❌ No paramétrico (p = {p_value:.4f})")

```

$$\sum_i^n x_i / n.$$

## Media Ponderada


La **media ponderada** es un tipo de promedio en el que cada valor tiene una importancia o peso diferente.

Si tenemos un conjunto de valores $( x_1, x_2, ..., x_n)$ con pesos correspondientes $(w_1, w_2, ..., w_n)$, la media ponderada se define como:

$
\bar{x}_p = \frac{w_1 x_1 + w_2 x_2 + \cdots + w_n x_n}{w_1 + w_2 + \cdots + w_n}
$

donde:

- $( x_i)$ son los valores.
- $(w_i)$ son los pesos asignados a cada valor.




```python
# Eliminar filas con valores nulos en las dos columnas relevantes
df_filtrado = df_datos_relacion[['ingreso', 'veces_mes_uso_td_compras_establecimientos']].dropna()

#Tamaño del dataframe
print('Tamaño del dataframe: \n', df_filtrado.shape)
print('---------------'*10)
print('Valores distintos de veces_mes_uso_td_compras_establecimientos: \n',df_filtrado.groupby('veces_mes_uso_td_compras_establecimientos').size())

# Calcular media ponderada
numerador = sum(i * p for i, p in zip(df_filtrado['ingreso'], df_filtrado['veces_mes_uso_td_compras_establecimientos']))
denominador = sum(df_filtrado['veces_mes_uso_td_compras_establecimientos'])

media_ponderada = numerador / denominador
print('---------------'*10)
print(f"Media ponderada del ingreso (ponderado por veces_mes_uso_td_compras_establecimientos): {media_ponderada:.2f}")


```

#Medidas de dispersión

## Varianza

La varianza es una medida estadística de la dispersión de un conjunto de datos respecto a su media. Indica cuánto se alejan los datos individuales del promedio; una varianza alta significa que los datos están muy dispersos, mientras que una varianza baja indica que los datos están más agrupados cerca de la media. Se calcula como el promedio de las desviaciones al cuadrado de cada dato respecto a la media.

### Poblacional

$$\sigma^2 = \frac {\sum(x_i - \mu)^2}{n} $$

### Muestral

$$s^2 = \frac {\sum(x_i - \bar{x})^2}{n - 1} $$

```python
#Muestral
varianza_edad = df_datos_relacion['edad'].var()
print(f"Varianza muestral de edad: {varianza_edad}")

#Poblacional
varianza_edad_poblacional = df_datos_relacion['edad'].var(ddof=0)
print(f"Varianza poblacional de edad: {varianza_edad_poblacional}")

#Muestral
varianza_edad_m_s = statistics.variance(df_datos_relacion['edad'])
print(f"Varianza muestral de edad con statistics: {varianza_edad_m_s}")

#Poblacional
varianza_edad_p_s = statistics.pvariance(df_datos_relacion['edad'])
print(f"Varianza poblacional de edad con statistics: {varianza_edad_p_s}")
```

## Desviación estándar

La desviación estándar es una medida estadística de la dispersión de un conjunto de datos, indicando qué tan cerca o lejos están los valores individuales de la media o promedio. Una desviación estándar baja significa que los datos son consistentes y se agrupan cerca de la media, mientras que una desviación estándar alta señala que los datos están más dispersos. Se calcula como la raíz cuadrada de la varianza.

### Poblacional

$$\sigma = \sqrt{ \frac {\sum (x_i - \mu)^2}{n} }$$

### Muestral

$$s = \sqrt \frac {\sum(x_i - \bar{x})^2}{n - 1} $$

```python
#Muestral
desviacion_estandar_edad = df_datos_relacion['edad'].std()
print(f"Varianza muestral de edad: {desviacion_estandar_edad}")

#Poblacional
desviacion_estandar_edad_poblacional = df_datos_relacion['edad'].std(ddof=0)
print(f"Varianza poblacional de edad: {desviacion_estandar_edad_poblacional}")

#Muestral
desviacion_estandar_edad_m_s = statistics.stdev(df_datos_relacion['edad'])
print(f"Varianza muestral de edad con statistics: {desviacion_estandar_edad_m_s}")

#Poblacional
desviacion_estandar_edad_p_s = statistics.pstdev(df_datos_relacion['edad'])
print(f"Varianza poblacional de edad con statistics: {desviacion_estandar_edad_p_s}")
```

## Sesgo
La asimetría mide cuánto se desvía la forma de la distribución respecto a una distribución normal (simétrica).

* Una distribución normal tiene asimetría = 0.
* Si los datos tienen cola hacia la derecha (más valores bajos y pocos valores muy altos), la asimetría será positiva.
* Si los datos tienen cola hacia la izquierda (más valores altos y pocos valores muy bajos), la asimetría será negativa.

```python
skew_val = df_datos_relacion['ingreso'].skew()

# Evaluar tipo de asimetría
if skew_val > 0:
    interpretacion = "Sesgo positivo (cola hacia la derecha)"
elif skew_val < 0:
    interpretacion = "Sesgo negativo (cola hacia la izquierda)"
else:
    interpretacion = "Distribución simétrica"

# Imprimir resultado
print(f"Asimetría (skew) de 'ingreso': {skew_val:.3f} → {interpretacion}")


```

# Correlación

La correlación es una medida estadística que indica si dos variables numéricas están relacionadas y de qué manera, evaluando la fuerza y dirección de esa asociación. Es importante entender que la correlación mide la dependencia conjunta de las variables, pero no implica una relación de causa y efecto.

### Tipos de Correlación
* Correlación Positiva o Directa (cercana a 1): Las dos variables tienden a aumentar o disminuir en el mismo sentido.
* Correlación Negativa o Inversa (cercana a -1): Una variable aumenta cuando la otra disminuye.
* Correlación Nula (cercana a 0): Indica que no hay una relación lineal significativa entre las variables.


```python
#Crear matriz de correlación
matriz_correlacion = df_datos_relacion.corr()

display(matriz_correlacion)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Mapa de calor
plt.figure(figsize=(30, 12))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Matriz de Correlación')
plt.show()
```

```python

#Obtener correlaciones a partir de moderadas > 0.5
# Convertir matriz a formato "long" para facilitar el filtrado
corr_unstacked = matriz_correlacion.unstack()

# Quitar correlaciones de la diagonal (correlación de variable consigo misma = 1)
corr_unstacked = corr_unstacked[corr_unstacked.index.get_level_values(0) != corr_unstacked.index.get_level_values(1)]

# Filtrar correlaciones con valor absoluto mayor o igual a 0.5
corr_moderada = corr_unstacked[abs(corr_unstacked) >= 0.5]

# Eliminar duplicados (porque la matriz es simétrica)
corr_moderada = corr_moderada.drop_duplicates()

# Ordenar de mayor a menor (por valor absoluto)
corr_moderada = corr_moderada.reindex(corr_moderada.abs().sort_values(ascending=False).index)

display(corr_moderada)


```

## Hipótesis

 $H_0$ = No hay correlación entre tipo atención médica y tiene cuenta nómina (correlación = 0).

 $H_a$ = Sí hay correlación (correlación ≠ 0).


```python
from scipy.stats import pearsonr

# Quitar nulos
df_pair = df_datos_relacion[['tipo_atencion_medica', 'tiene_cuenta_nomina']].dropna()

#Tamaño del dataframe
print('Tamaño del dataframe: \n', df_pair.shape)

# Realizar test de Pearson
corr_coef, p_val = pearsonr(df_pair['tipo_atencion_medica'], df_pair['tiene_cuenta_nomina'])

print(f"Coeficiente de correlación Pearson: {corr_coef:.3f}")
print(f"Valor p: {p_val:.4f}")

if p_val < 0.05:
    print("Rechazamos H0: Por lo tanto hay correlación significativa entre tipo_atencion_medica y tiene_cuenta_nomina")
else:
    print("No se rechaza H0: No hay evidencia suficiente de correlación")
```
