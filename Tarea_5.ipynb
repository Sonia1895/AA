##Verificar cual es el mejor valor para PCA

El PCA es una técnica lineal de reducción de dimensionalidad que:

1. Centra y escala los datos (media 0, varianza 1).

2. Calcula la matriz de covarianzas de las variables.

3. Obtiene los vectores y valores propios (eigenvectors/eigenvalues) de esa matriz.

4. Ordena los componentes según la cantidad de varianza explicada (eigenvalues).

Cada componente principal es una combinación lineal de las variables originales:

$$𝑍_𝑘= 𝑎_{𝑘1}x_1 + 𝑎_{𝑘2}x_2 +...+𝑎_{𝑘p}x_p $$

Donde:
* $Z_k$: el k-ésimo componente principal
* $𝑎_{𝑘j}$: epeso del eje j en el componente k
* $x_j$: variable original

En términos prácticos:

El primer componente (PC1) explica la mayor varianza.

El segundo (PC2) explica la siguiente mayor, y así sucesivamente.

```python
#Seleccionar componentes de acuerdo a varianza acumulada, contempland todos los datos
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

datos = df_datos_relacion_na.drop(columns=["tiene_tdc_credito", "llavemod"]).copy()  # quitamos del dato de llave y tiene tdc

# Escalado
scaler = StandardScaler()
datos_scaled = scaler.fit_transform(datos)

# PCA completo
pca = PCA()
X_pca = pca.fit_transform(datos_scaled)

# Varianza explicada por componente
varianza_explicada = pca.explained_variance_ratio_

# Varianza acumulada
varianza_acumulada = varianza_explicada.cumsum()

# Gráfico
plt.figure(figsize=(10,6))
plt.plot(range(1, len(varianza_acumulada)+1), varianza_acumulada, marker='o')
plt.xlabel("Número de componentes")
plt.ylabel("Varianza acumulada")
plt.title("Curva de varianza acumulada PCA")
plt.axhline(y=0.9, color='r', linestyle='--', label='90% de varianza')
plt.legend()
plt.grid(True)
plt.show()

# Selección de componentes
n_components = next(i for i, var in enumerate(varianza_acumulada) if var >= 0.9) + 1
print(f"Número de componentes para ~90% de varianza: {n_components}")
```

º##Existen métricas de validación de clustering que ayudan a elegir el número de grupos o a comparar diferentes resultados

| Estrategia                       | Descripción                                                                      | Aplicable a           |
| -------------------------------- | -------------------------------------------------------------------------------- | --------------------- |
| **Silhouette Score**             | Compara distancia intra-cluster vs inter-cluster → entre -1 y 1                  | KMeans, DBSCAN        |
| **Davies-Bouldin Index (DB)**    | Relación entre dispersión de clusters y distancia entre clusters → menor = mejor | KMeans, DBSCAN        |

* Silhouette Score → mide cómo de coherentes son los clusters frente a los outliers

* Calinski-Harabasz → evalúa densidad y separación de clusters



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt


X_onehot = df_onehot.drop(columns=["tiene_tdc_credito","llavemod"]).copy()
X_onehot_scaled = scaler.fit_transform(X_onehot)


# --- PCA para reducción de dimensionalidad ---
pca = PCA(n_components=61)  # de acuerdo al punto anterior
X_pca = pca.fit_transform(X_onehot_scaled)

# --- Rango de parámetros DBSCAN ---
eps_values = [0.75,1.0,1.5, 2.0]
min_samples_values = [3,5,8,10]

results = []

# --- Prueba automática de parámetros y métricas ---
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(X_pca)

        # Filtrar outliers (-1) para métricas
        mask = clusters != -1
        unique_clusters = len(set(clusters[mask]))
        if np.sum(mask) > 1 and unique_clusters > 1:  # al menos 2 clusters
            sil_score = silhouette_score(X_pca[mask], clusters[mask])
            dbi_score = davies_bouldin_score(X_pca[mask], clusters[mask])
        else:
            sil_score = -1
            dbi_score = np.inf  # Davies-Bouldin alto indica mala calidad

        n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
        n_outliers = np.sum(clusters == -1)

        results.append({
            "eps": eps,
            "min_samples": min_samples,
            "n_clusters": n_clusters,
            "n_outliers": n_outliers,
            "silhouette": sil_score,
            "davies_bouldin": dbi_score
        })

# Convertir a dataframe y ordenar por mejor Silhouette (mayor) y Davies-Bouldin (menor)
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by=["silhouette", "davies_bouldin"], ascending=[False, True])
print("=== Resultados métricas DBSCAN ===")
print(results_df)

# --- Selección de los mejores parámetros ---
best_params = results_df.iloc[0]
best_eps = best_params["eps"]
best_min_samples = int(best_params["min_samples"])
print(f"\nMejores parámetros -> eps: {best_eps}, min_samples: {best_min_samples}")

# --- 6DBSCAN final con mejores parámetros ---
dbscan_final = DBSCAN(eps=best_eps, min_samples=best_min_samples)
clusters_final = dbscan_final.fit_predict(X_pca)


# --- Conteo clusters y outliers ---
print("\n=== Conteo clusters finales ===")
print(pd.Series(clusters_final).value_counts())

# --- Visualización en 2D ---
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X_onehot_scaled)

plt.figure(figsize=(10,6))
plt.scatter(X_2d[:,0], X_2d[:,1], c=clusters_final, cmap='tab20', s=50)
plt.title(f"DBSCAN final con PCA 60 -> eps={best_eps}, min_samples={best_min_samples}")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label="Cluster")
plt.show()
```
