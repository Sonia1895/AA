# Aplicar un método de filtro: SelectKBest

SelectKBest es una técnica que selecciona las K características más importantes basándose en una prueba estadística.

```python
import pandas as pd

#Definir quien es x y y
X = df_datos_relacion_na.drop(columns=["tiene_tdc_credito", "llavemod"])  # quitamos del dato de llave y tiene tdc
y = df_datos_relacion_na["tiene_tdc_credito"] #definimo y

display(X.head())
display(y.head())

```

```python
#Código para seleccionar el valor de K
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Supongo que ya tienes X, y definidos

# 1. Escalar las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Definir rango de K a evaluar
ks = range(5, 30)
scores = []

for k in ks:
    # Selección de las mejores k características
    selector = SelectKBest(score_func=f_classif, k=k)
    X_new = selector.fit_transform(X_scaled, y)

    # Modelo con configuración que asegura convergencia
    model = LogisticRegression(solver='saga', max_iter=7000, tol=1e-3)

    # Validación cruzada con 5 folds y scoring por precisión
    cv_score = cross_val_score(model, X_new, y, cv=5, scoring='accuracy').mean()
    scores.append(cv_score)

# 3. Graficar resultados
plt.plot(ks, scores, marker='o')
plt.xticks(range(min(ks), max(ks) + 1, 1))
plt.xlabel('Número de características (K)')
plt.ylabel('Precisión promedio (CV)')
plt.title('Selección óptima de K con LogisticRegression (solver=saga)')
plt.grid(True)
plt.show()

```

**¿Qué hace f_classif?**

f_classif realiza un test ANOVA F (Análisis de Varianza) para medir la relación entre cada característica (feature) y la variable objetivo en un problema de clasificación.

* Para cada característica, calcula un valor F que indica qué tan bien esa característica diferencia entre las clases.

* Cuanto más alto sea el valor F, más importante es la característica para distinguir entre las clases.

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import LabelEncoder
import pandas as pd


# Selección de características
selector = SelectKBest(score_func=f_classif, k=16)
selector.fit(X, y)

# Mostrar las variables seleccionadas
selected_features = X.columns[selector.get_support()]
display("Características seleccionadas:", selected_features.tolist())

feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'F_score': f_scores
}).sort_values(by='F_score', ascending=False)


plt.figure(figsize=(12, 6))
plt.barh(feature_scores['Feature'], feature_scores['F_score'], color='skyblue')
plt.xlabel('F-score')
plt.title('Importancia de cada variable según f_classif')
plt.gca().invert_yaxis()  # Poner la más importante arriba
plt.grid(axis='x')
plt.tight_layout()
plt.show()

```

**¿Qué hace f_regression?**

* Pruebas de regresión lineal univariada que devuelven valores p y estadísticos F.

* Modelo lineal rápido para probar el efecto de un solo regresor.


```python
from sklearn.feature_selection import f_regression

valor_f = f_regression(X, y)

display(valor_f)

pasan_prueba = []
no_pasan_prueba = []
alpha = 0.05

for i in range(len(valor_f[1])):
    if valor_f[1][i] < alpha:
        pasan_prueba.append(X.columns[i])
    else:
        no_pasan_prueba.append(X.columns[i])

display("Características que pasan la prueba:", pasan_prueba)
display("Características que no pasan la prueba:", no_pasan_prueba)

df_resultados = pd.DataFrame(valor_f[1], index = X.columns, columns=['valor_f'])
display(df_resultados)

colores = []
for f in df_resultados.index:
    if f in pasan_prueba:
        colores.append('g')
    else:
        colores.append('r')

plt.figure()
plt.bar(df_resultados.index, df_resultados.valor_f, color = colores)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Escalar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Modelo base
model = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)

# Parámetros a probar (valores de C)
param_grid = {'C': np.logspace(-4, 2, 20)}  # De 0.0001 a 100, 20 valores log espaciales

# Validación cruzada estratificada para clasificación
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Grid Search para encontrar el mejor C (máxima accuracy)
grid_search = GridSearchCV(model, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_scaled, y)

print(f"Mejor valor de C: {grid_search.best_params_['C']}")
print(f"Mejor accuracy: {grid_search.best_score_:.4f}")

# Entrenar modelo final con mejor C
best_model = grid_search.best_estimator_

# Obtener coeficientes del mejor modelo
coef = best_model.coef_[0]
coef_df = pd.DataFrame({'feature': X.columns, 'coefficient': coef})

# Variables seleccionadas (coeficiente != 0)
selected_features = coef_df[coef_df['coefficient'] != 0]['feature'].tolist()
removed_features = coef_df[coef_df['coefficient'] == 0]['feature'].tolist()

print(f"Variables seleccionadas ({len(selected_features)}): {selected_features}")
print(f"Variables eliminadas ({len(removed_features)}): {removed_features}")
```

# Árboles de decisión (Random Forest)

Usa la importancia de cada característica basada en ganancia de información.

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

# 1. Entrenar Random Forest para obtener importancias
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X, y)

# 2. Ordenar importancias
importances = pd.Series(rf.feature_importances_, index=X.columns)
sorted_features = importances.sort_values(ascending=False).index.tolist()

# 3. Evaluar rendimiento con distintas cantidades de top features
results = {}

for n in range(5, 30):  # Por ejemplo, de 5 a 30 features
    selected_features = sorted_features[:n]
    X_subset = X[selected_features]

    # Validación cruzada (5 folds, puedes ajustar)
    scores = cross_val_score(RandomForestClassifier(n_estimators=100),
                             X_subset, y, cv=5, scoring='accuracy')

    results[n] = scores.mean()

# 4. Encontrar el número óptimo de features
best_n = max(results, key=results.get)
print(f"Mejor cantidad de features: {best_n} con accuracy promedio: {results[best_n]:.4f}")


display("Top características por Random Forest:", sorted_features)


plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('Número de features')
plt.ylabel('Accuracy promedio (cross-validation)')
plt.title('Selección del número óptimo de features')
plt.grid(True)
plt.show()

```

## EFS

EFS hace referencia a: Exhaustive Feature Selector de la librería mlxtend.feature_selection.

Este selector prueba todas las combinaciones posibles de características entre min_features y max_features, evaluando cada una con la métrica especificada mediante validación cruzada.

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression


lr = LinearRegression()

efs = EFS(
    estimator = lr,                      # Usa una regresión lineal como modelo base (estimator)
    min_features = 5,                    # Considera combinaciones de al menos 1 característica
    max_features = 10,                    # Hasta un máximo de 10 características por combinación
    scoring = 'neg_mean_absolute_error', # Métrica: error absoluto medio negativo (para regresión)
    cv = 5                               # Validación cruzada con 5 particiones (folds)
)


efs.fit(X, y)

print('Mejor combinación de features:', efs.best_feature_names_)
print('Puntaje del mejor subset:', efs.best_score_)

df_resultados_efs = pd.DataFrame(efs.get_metric_dict()).T
df_resultados_efs = df_resultados_efs.sort_values('avg_score', ascending = False)
display(df_resultados_efs)

plt.figure()

plt.plot(
    df_resultados_efs.iloc[:5].feature_names.astype(str),
    df_resultados_efs.iloc[:5].avg_score,
)

plt.xticks(rotation = 45)

plt.plot()
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS


sfs = SFS(estimator = lr,       # Usa una regresión lineal como modelo base (estimator)
          k_features = (5, 30),  # Considera alguna combina enre 5 y 10 variables
          forward = False,       # Realiza selección hacia atrás (backward)
          scoring = 'neg_mean_absolute_error', # Métrica para evaluar el modelo
          cv=5)                  # Validación cruzada con 5 particiones

sfs = sfs.fit(X, y)

pd.DataFrame(sfs.subsets_).T.sort_values('avg_score', ascending = False)

print('Mejor combinación de features:', sfs.k_feature_names_)
print('Puntaje del mejor subset:', sfs.k_score_)


df_resultados_sfs = pd.DataFrame(sfs.get_metric_dict()).T
df_resultados_sfs = df_resultados_sfs.sort_values('avg_score', ascending = False)
display(df_resultados_sfs)


plt.figure()

plt.plot(
    df_resultados_sfs.iloc[:5].feature_names.astype(str),
    df_resultados_sfs.iloc[:5].avg_score,
)

plt.xticks(rotation = 45)

plt.plot()
```

# Investigación métricas para seleccionar características

[Selección de Variables en Machine Learning con Python](https://cienciadedatos.net/documentos/py67-seleccion-variables-machine-learning-python#Instrucciones_para_citar)

* [f_regression:](https://qu4nt.github.io/sklearn-doc-es/modules/generated/sklearn.feature_selection.f_regression.html) Modelo lineal para probar el efecto individual de cada uno de los muchos regresores. Esta es una función de puntuación para ser utilizada en un procedimiento de selección de características, no un procedimiento de selección de características autónomo.

Esto se hace en 2 pasos:

Se calcula la correlación entre cada regresor y el objetivo.

Se convierte a una puntuación F y luego a un valor-p.

* [f_classif](https://machinelearningmastery.com/feature-selection-machine-learning-python/) realiza un test ANOVA F (Análisis de Varianza) para medir la relación entre cada característica (feature) y la variable objetivo en un problema de clasificación.

Para cada característica, calcula un valor F que indica qué tan bien esa característica diferencia entre las clases.

Cuanto más alto sea el valor F, más importante es la característica para distinguir entre las clases.

* [La regularización L1, o regresión lazo](https://builtin.com/data-science/l2-regularization) (operador de selección y contracción mínima absoluta), es un método de regularización que penaliza los coeficientes de alto valor en un modelo de aprendizaje automático. El término de penalización de L1 es el valor absoluto de la suma de coeficientes (o el valor absoluto de la magnitud del coeficiente) y se suma a la función de pérdida de la suma de errores cuadrados (SSE) del modelo.

La regresión Lasso puede reducir los valores de los coeficientes a cero, lo que permite la selección y eliminación de características . Este aspecto también permite que la regresión Lasso gestione cierta multicolinealidad (altas correlaciones entre características) en un conjunto de datos sin afectar la interpretabilidad. Sin embargo, la regresión Lasso no es adecuada para una alta multicolinealidad: si existen covariables altamente correlacionadas, la regularización L1 eliminará aleatoriamente una de las características del modelo.

* [Random Forest](https://www.geeksforgeeks.org/machine-learning/feature-selection-using-random-forest/)  es un algoritmo versátil de aprendizaje automático que funciona construyendo múltiples árboles de decisión durante el entrenamiento y generando la moda de las clases (clasificación) o la predicción de la media (regresión) de cada árbol. Combina los conceptos de bagging (agregación bootstrap) y selección aleatoria de características, lo que mejora la precisión y la robustez.

Random Forest es especialmente adecuado para la selección de características por varias razones:

Clasificación de características intrínsecas: Random Forest proporciona un método integrado para evaluar la importancia de las características.

Maneja alta dimensionalidad: eficaz incluso cuando el número de características es mucho mayor que el número de muestras.

No linealidad: puede capturar interacciones complejas entre características sin requerir una especificación explícita de las interacciones.

* La selección secuencial de características ([SFS](https://www.geeksforgeeks.org/machine-learning/sequential-feature-selection/)) es un algoritmo voraz que añade o elimina iterativamente características de un conjunto de datos para mejorar el rendimiento de un modelo predictivo. La SFS puede ser de selección progresiva o regresiva.

Las ventajas de la selección secuencial de características incluyen:

Es un algoritmo simple y eficiente.
Se puede utilizar con cualquier tipo de modelo predictivo.
Se puede utilizar para seleccionar características para tareas de clasificación y regresión.

Las desventajas de la selección secuencial de características incluyen:

Puede ser sensible a la elección de la métrica de puntuación.
Puede estar sesgado hacia características que están altamente correlacionadas con la característica objetivo.
Puede resultar computacionalmente costoso para conjuntos de datos grandes.
